```python
import pandas as pd

result = pd.DataFrame({'col': [1, 2, 3]})
expected = pd.DataFrame({'col': [1, 2, 3]})

# Try the wrong way
print("result == expected:")
print(result == expected)  # DataFrame of bools!

# Try to assert (this will fail)
try:
    assert result == expected
except ValueError as e:
    print(f"Error: {e}")
```
